```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2], [3, 4]])
print(array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print(random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeroes_array = np.zeros((3, 2))
print(ones_array, zeroes_array)
```

```python
arr_6x3 = np.random.rand(6, 3)
transposed_arr = arr_6x3.T  # or np.transpose(arr_6x3)
print(transposed_arr.shape) # Result will be (3, 6)
```

```python
n, p, m = 3, 4, 2
matrix_A = np.random.rand(n, p)
matrix_B = np.random.rand(p, m)

dot_product = np.dot(matrix_A, matrix_B)
# Alternatively: dot_product = matrix_A @ matrix_B
print(dot_product.shape) # Result will be (n, m)
```
